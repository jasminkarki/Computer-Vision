*Pre trained Model*

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
import skimage 

test_image = skimage.data.astronaut()
test_image.shape
```

```python
plt.imshow(test_image)
```

*3 channel image(RGB) to 2 channel(Black and White)*

```python
# plt.imshow(test_image, cmap='gray')
```

```python
face_image=test_image
```

```python
def detect_face(img): 
  face_rect=face_cascade.detectMultiScale(face_image,scaleFactor=1.2,minNeighbors=5)  #Return coordinates of the rectangular image of the face detected (Rectangle drawn over face)
  print(face_rect)

  for (x,y,w,h) in face_rect:
    cv2.rectangle(face_image,(x,y),(x+w,y+h),(255,255,255),10)

  return face_image
```

```python
returned_image= detect_face(face_image)
plt.imshow(returned_image)
```

Because of one face..only one coordinate has been returned
<br>
x, y, width, height is returned<br>
Group Photos should return coordinate=number of face

```python
eyes_image=test_image
```

```python
def detect_eyes(img): 
  eye_rect=eye_cascade.detectMultiScale(eyes_image,scaleFactor=1.2,minNeighbors=5)
  print(eye_rect)

  for (x, y, w, h) in eye_rect:
    cv2.rectangle(eyes_image,(x,y),(x+w,y+h),(255,255,255),4)
  return eyes_image
```

```python
returned_eyes=detect_eyes(eyes_image)
plt.imshow(returned_eyes)
```
